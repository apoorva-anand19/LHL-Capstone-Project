```python

from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from statsmodels.stats.outliers_influence import variance_inflation_factor

import pandas as pd
import numpy as np
import statsmodels.api as sm

filtered_df = pd.read_csv('C:\\Users\\apoor\\OneDrive\\Documents\\Data Analytics LHL\\LHL-Capstone-Project\\CSVs\\Filtered_gamesData.csv')
filtered_df
```

```python
filtered_df = filtered_df.drop(['visitor_team_wins','season'], axis=1)
filtered_df
```

```python
more_filtered_df = filtered_df.drop(['visitor_team_score'], axis=1)
more_filtered_df
```

```python
x = more_filtered_df.drop('home_team_wins', axis=1)
y = more_filtered_df['home_team_wins']

x = sm.add_constant(x)
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)
```

```python
model = sm.Logit(y_train, x_train)

result = model.fit()
print(result.summary())
```

- This gives us an obvious answer of, more chances of wining for a team if they score more.
